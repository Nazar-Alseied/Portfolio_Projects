```python
#importing liberaries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#reading the data
df = pd.read_csv('/content/movies.csv')
```

```python
#let us look at the data
df
```

```python
df.dtypes
```

```python
# cheking for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# cheking outliers in groos column
df.boxplot(column=['gross'])
```

```python
# removing duplicated data
df.drop_duplicates()
```

```python
#  top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
# regretion show relation ship between gross and budget
sns.regplot(x="gross", y="budget", data=df ,scatter_kws={"color":"blue"}, line_kws={"color":"green"})
```

```python
# correlation 
df.corr(method ='pearson')
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation of Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```
